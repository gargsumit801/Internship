### Q1. --> Write a python program to display all the header tags from wikipedia.org. 

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Main_Page'
```

```python
page = requests.get(url)
```

```python
page
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
heading_tags =['h1','h2']
```

```python
for tags in soup.find_all(heading_tags):
    print(tags.name + ' -> ' + tags.text.strip())
```

### Q.2)	Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame. 

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/list/ls091520106/')
page
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
movie_name = soup.find('h3',class_="lister-item-header")
```

```python
movie_list = []
for i in soup.find_all('h3',class_="lister-item-header"):
    movie_list.append(i.text.replace('\n',''))
movie_list
```

```python
year = []
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
year
```

```python
Rating = []
for i in soup.find_all('span',class_="ipl-rating-star__rating"):
    Rating.append(i.text)
    
Rating
```

### Q.3)	Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL. 

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

### extract Name

```python
restra_name = soup.find('div',class_="restnt-info cursor")
```

```python
restra_name.text
```

### Extract location

```python
location = soup.find('div',class_="restnt-loc ellipsis")
```

```python
location.text
```

### Extract rating

```python
rating = soup.find('div',class_="restnt-rating rating-4")
```

```python
rating.text
```

### Image URL

```python
images = []
for i in soup.find_all('img',class_="no-img"):
    images.append(i['data-src'])
images
```

### Cuisine

```python
cuisines = soup.find_all('span', class_='double-line-ellipsis')
```

```python
cuisine_list = []
```

```python
for cuisine in cuisines:
    cuisine_list.append(cuisine.text.strip())
```

```python
cuisine_list
```
