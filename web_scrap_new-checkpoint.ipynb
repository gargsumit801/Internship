b) Top 10 ODI Batsmen along with the records of their team and rating.

c) Top 10 ODI bowlers along with the records of their team and rating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
web = requests.get('https://www.icc-cricket.com/rankings/team-rankings/mens/odi')
```

```python
soup = BeautifulSoup(web.text,'html.parser')
```

```python
teams = soup.find_all('span', class_='u-hide-phablet')
matches = soup.find_all('td', class_='rankings-block__banner--matches')
points = soup.find_all('td', class_='rankings-block__banner--points')
ratings = soup.find_all('td', class_='rankings-block__banner--rating')
```

```python
print("Top 10 ODI Teams:")
print("{:<10} {:<10} {:<10} {:<10}".format("teams", "matches", "points", "rating"))
```

```python
def scrape_top_10_teams():
    url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    teams = soup.find_all('span', class_='u-hide-phablet')
    matches = soup.find_all('td', class_='rankings-block__banner--matches')
    points = soup.find_all('td', class_='rankings-block__banner--points')
    ratings = soup.find_all('td', class_='rankings-block__banner--rating')

    print("Top 10 ODI Teams:")
    print("{:<30} {:<10} {:<10} {:<10}".format("Team", "Matches", "Points", "Rating"))
    for i in range(10):
        team = teams[i].text
        match = matches[i].text
        point = points[i].text
        rating = ratings[i].text.strip()
        print("{:<30} {:<10} {:<10} {:<10}".format(team, match, point, rating))
```

## Q 1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data

```python
web = requests.get('https://www.imdb.com/list/ls056092300/')
```

```python
soup = BeautifulSoup(web.content)
```

```python
movie = soup.find('h3',class_="lister-item-header")
```

```python
movie.text.strip()
```

```python
movie_name = []
for i in soup.find_all('h3',class_="lister-item-header"):
    movie_name.append(i.text.strip().replace('\n',''))
```

```python
movie_name
```

```python
year = []
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
```

```python
year
```

```python
rate =[]
for i in soup.find_all('div',class_="ipl-rating-star small"):
    rate.append(i.text.strip())
```

```python
rate
```

```python
df = pd.DataFrame({'movie_name':movie_name,'year':year,'rate':rate})
```

## 6) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .

```python
page = requests.get('https://www.bewakoof.com/bestseller?sort=popular')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
product_name = []
for i in soup.find_all('h2',class_='clr-shade4'):
    product_name.append(i.text)
```

```python
product_name
```

```python
price = []
for i in soup.find_all('div',class_="discountedPriceText"):
    price.append(i.text)

price
```

```python
image_link = []
for i in soup.find_all('img',class_="productImgTag"):
    image_link.append(i['src'])
    
image_link
```

```python
df = pd.DataFrame({'product_name':product_name,'price':price,'image_link':image_link})
df
```
