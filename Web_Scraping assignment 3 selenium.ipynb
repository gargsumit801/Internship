### Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.

The above task will be done in following steps:
    
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap a)Quote b) Author c) Type Of Quotes

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.azquotes.com/')
```

```python
top = driver.find_element(By.XPATH,'//div[@class="mainmenu"]/ul/li[5]/a')
top.click()
```

```python
Qoutes = []
for page in range(0,10):
    a = driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')
    for i in a:
        Qoutes.append(i.text)
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]/a')
    next_button.click()
    time.sleep(3)
```

```python
Qoutes
```

```python
len(Qoutes)
```

```python
Author = []
for page in range(0,10):
    b = driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for i in b:
        Author.append(i.text)
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]/a')
    next_button.click()
    time.sleep(3) 
```

```python
Author
```

```python
len(Author)
```

```python
type_of_Qoutes = []
for page in range(0,10):
    c = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in c:
        type_of_Qoutes.append(i.text)
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]/a')
    next_button.click()
    time.sleep(3)
```

```python
type_of_Qoutes
```

```python
len(type_of_Qoutes)
```

```python
df = pd.DataFrame({'Qoutes':Qoutes,'Author':Author,'type_of_Qoutes':type_of_Qoutes})
df
```
